2.Write a program to plot following equation over time. the equation defines a dynamical system in which position of the system changes with time . you have to plot the positions over time. Think of this as a Bee moving in 3d space. you need to plot the path taken by the Bee.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the system of differential equations
def system(y, t, a, b, c):
    x, y, z = y
    dxdt = a * (y - b)
    dydt = b * x - y - x * z
    dzdt = x * y - c * z
    return [dxdt, dydt, dzdt]

# Define parameters
a = 10
b = 28
c = 2.667

# Initial conditions
initial_conditions = [0, 1, 1.05]  # Initial values for x, y, and z

# Time points for plotting
t = np.linspace(0, 50, 5000)  # Create a time vector from 0 to 50 with 5000 points

# Solve the system of differential equations
solution = odeint(system, initial_conditions, t, args=(a, b, c))

# Extract the solution components
x, y, z = solution.T

# Plot the solutions
plt.figure(figsize=(12, 6))
plt.subplot(311)
plt.plot(t, x, label='x')
plt.xlabel('Time')
plt.ylabel('x')
plt.legend()

plt.subplot(312)
plt.plot(t, y, label='y')
plt.xlabel('Time')
plt.ylabel('y')
plt.legend()

plt.subplot(313)
plt.plot(t, z, label='z')
plt.xlabel('Time')
plt.ylabel('z')
plt.legend()

plt.tight_layout()
plt.show()
```
